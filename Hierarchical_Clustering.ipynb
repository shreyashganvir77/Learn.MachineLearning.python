Importing the Librabries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importing the Dataset.

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
```

Using Dendrogram for number of CLusters.

```python
from scipy.cluster.hierarchy import dendrogram, linkage
dendrogram = dendrogram(linkage(x, method = 'ward'))
plt.title('Dendrogram')
plt.show()
```

Training the Hierarchical Model on the whole Dataset.

```python
# In this typr of learning each point is considered as a cluster in the begining i.e there are N Clusters.
# Next, the point which is closer to first made cluster is inclued in that cluster and we N-1 Clusters.
# Now, we select two clusters which are close to each other and form one cluster/
# ..... there can be many ways to define the closest path, here we consider the distance between the centroids of the cluster.
# This process is repeated until there is one cluster.
# To select no of clusters in out dataset, we use dendrogram for this. 
# In the above dendrogram the longest distance between two horizontal bars is selected and then horizontal line is cut through it.
# The no points the horizontal line cuts the dendrogram is the no of clusters we need in our dataset. 

from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_pred = clustering.fit_predict(x)
print(y_pred)
```

Visualizing the Results.

```python
# For Visualizing the results we plot the clusters using the y_pred array for selecting the particular customers in the Dataset.

plt.scatter(x[y_pred==0, 0], x[y_pred==0, 1], label = 'Cluster1')
plt.scatter(x[y_pred==1, 0], x[y_pred==1, 1], label = 'Cluster2')
plt.scatter(x[y_pred==2, 0], x[y_pred==2, 1], label = 'Cluster3')
plt.scatter(x[y_pred==3, 0], x[y_pred==3, 1], label = 'Cluster4')
plt.scatter(x[y_pred==4, 0], x[y_pred==4, 1], label = 'Cluster5')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.title('Clustering of Customers')
plt.legend()
plt.show()
```
